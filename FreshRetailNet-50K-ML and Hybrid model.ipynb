## Prepare Data for ML Models

### Subtask:
Convert the 'dt' column to datetime objects in both `df_final` and `df_final_eval`. Identify categorical and numerical features from the remaining columns. Apply label encoding to the identified categorical features for compatibility with models like XGBoost. Separate the data into training (from `df_final`) and testing (from `df_final_eval`) sets, defining features (X) and the target variable (y = 'sale_amount_log').


```python
df_final['dt'] = pd.to_datetime(df_final['dt'])
df_final_eval['dt'] = pd.to_datetime(df_final_eval['dt'])

print("Converted 'dt' column to datetime objects in df_final and df_final_eval.")
print("df_final['dt'] dtype:", df_final['dt'].dtype)
print("df_final_eval['dt'] dtype:", df_final_eval['dt'].dtype)
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns (excluding 'dt' and 'sale_amount_log')
categorical_cols = [
    'store_id', 'management_group_id', 'first_category_id',
    'second_category_id', 'third_category_id', 'product_id',
    'holiday_flag', 'activity_flag',
     'binary_stockout','is_month_start','is_month_end'
]

# Apply Label Encoding to categorical features in both dataframes
for col in categorical_cols:
    le = LabelEncoder()
    df_final[col] = le.fit_transform(df_final[col])
    df_final_eval[col] = le.transform(df_final_eval[col]) # Use transform for eval set to avoid new categories

print("Label Encoding applied to categorical columns in df_final and df_final_eval.")

# Define target variables
y_train = df_final['sale_amount_log']
y_test = df_final_eval['sale_amount_log']

# Define feature sets by dropping 'dt' and 'sale_amount_log'
X_train = df_final.drop(columns=['dt', 'sale_amount_log'])
X_test = df_final_eval.drop(columns=['dt', 'sale_amount_log'])

print("Target variables (y_train, y_test) and feature sets (X_train, X_test) created.")
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

## Train and Evaluate XGBoost Model

### Subtask:
Train an XGBoost Regressor model using the prepared training data. Make predictions on the test set. Inverse transform both the predictions and the actual 'sale_amount_log' to their original scale using `np.expm1`. Evaluate the model's performance using Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and R-squared (R2) on these inverse-transformed values. Visualize a sample of actual vs. predicted sales on the original scale.


```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Instantiate an XGBRegressor model
xgb_model = xgb.XGBRegressor(
    n_estimators=1000, # Number of boosting rounds
    learning_rate=0.05, # Step size shrinkage to prevent overfitting
    max_depth=7,        # Maximum depth of a tree
    subsample=0.7,      # Subsample ratio of the training instance
    colsample_bytree=0.7, # Subsample ratio of columns when constructing each tree
    random_state=42,    # Random seed for reproducibility
    n_jobs=-1,          # Use all available cores
    tree_method='hist'  # For faster training on large datasets
)

print("XGBoost Regressor model instantiated.")

# 3. Train the XGBoost model
xgb_model.fit(X_train, y_train)
print("XGBoost model trained.")

# 4. Generate predictions on the test set
y_pred_xgb_log = xgb_model.predict(X_test)
print("Predictions generated on the test set.")

# 5. Apply np.expm1 to inverse transform predictions and actual values
y_test_original = np.expm1(y_test)
y_pred_xgb_original = np.expm1(y_pred_xgb_log)
print("Predictions and actual values inverse transformed to original scale.")

# 6. Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_xgb_original))

# 7. Calculate the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test_original, y_pred_xgb_original)

# 8. Calculate the R-squared (R2) score
r2 = r2_score(y_test_original, y_pred_xgb_original)

# 9. Print the calculated RMSE, MAE, and R2 scores
print(f"\n--- XGBoost Model Performance on Original Scale ---")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R2 Score: {r2:.4f}")

# 10. Visualize a sample of actual vs. predicted sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_original.head(200), y=y_pred_xgb_original[:200], alpha=0.6)
plt.plot([y_test_original.min(), y_test_original.max()], [y_test_original.min(), y_test_original.max()], 'r--', label='Perfect Prediction')
plt.title('XGBoost: Actual vs. Predicted Sales (Original Scale - Sample)')
plt.xlabel('Actual Sales (Original Scale)')
plt.ylabel('Predicted Sales (Original Scale)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import optuna
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd # Import pandas for Series operations if needed

# Ensure categorical_cols and cat_features_indices are defined from previous steps
# X_train, y_train_residuals, X_test, y_test, prophet_predictions_for_test
# are assumed to be available from the global scope/previous cells.

def objective(trial):
    # 1. Suggest hyperparameters for CatBoostRegressor
    iterations = trial.suggest_int('iterations', 500, 2000) # Number of boosting rounds
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.2, log=True) # Step size shrinkage
    depth = trial.suggest_int('depth', 4, 10) # Maximum depth of a tree
    l2_leaf_reg = trial.suggest_float('l2_leaf_reg', 1e-3, 10.0, log=True) # L2 regularization term

    # 2. Instantiate a CatBoostRegressor with suggested parameters
    cat_residual_model_tuned = CatBoostRegressor(
        iterations=iterations,
        learning_rate=learning_rate,
        depth=depth,
        l2_leaf_reg=l2_leaf_reg,
        random_seed=42,
        verbose=0,       # Suppress verbose output
        cat_features=cat_features_indices
    )

    # 3. Train the CatBoost model on X_train and y_train_residuals
    cat_residual_model_tuned.fit(X_train, y_train_residuals)

    # 4. Predict residuals for the test set
    y_pred_cat_residuals_tuned = cat_residual_model_tuned.predict(X_test)

    # 5. Combine predicted residuals with Prophet forecasts to get final hybrid predictions
    # Ensure prophet_predictions_for_test is aligned with X_test. It was created by mapping dt from df_final_eval.
    y_pred_hybrid_log_tuned = prophet_predictions_for_test + y_pred_cat_residuals_tuned

    # 6. Inverse transform to original scale
    y_pred_hybrid_original_tuned = np.expm1(y_pred_hybrid_log_tuned)
    y_test_original_tuned = np.expm1(y_test) # y_test is already aligned to df_final_eval

    # Handle potential negative predictions after expm1, setting them to 0
    y_pred_hybrid_original_tuned[y_pred_hybrid_original_tuned < 0] = 0

    # 7. Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_test_original_tuned, y_pred_hybrid_original_tuned))

    return rmse

print("Optuna objective function 'objective' defined.")
```

```python
import optuna

# Create an Optuna study object and optimize the objective function
# 'minimize' is used because we want to minimize the RMSE
# n_trials specifies the number of optimization trials
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50) # You can adjust n_trials based on computational resources and desired exploration

print("Optuna optimization complete.")

# Print the best trial's parameters and value
print("\nBest trial:")
print(f"  Value (RMSE): {study.best_value:.4f}")
print("  Params: ")
for key, value in study.best_params.items():
    print(f"    {key}: {value}")
```
